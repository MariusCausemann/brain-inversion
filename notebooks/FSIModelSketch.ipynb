## Model sketch

* coupling of Biot (for the brain parenchyma) and Navier-Stokes (for the CSF flow in the SAS and ventricles)
* net blood inflow as pulsatile driving force (equally distributed within the brain and maybe the choroid plexus?)
* appropriate interface conditions (continuity of normal flux, momentum conservation, balance of fluid normal stresses...)
* no-slip BC at the skull for the fluid region, Neumann BC for outflow of CSF into the spinal cord



## what I hope to see:

* ICP amplitudes
* ICP gradient from ventricle to SAS
* morphology of the ICP curve (Subpeaks P1, P2, P3), espacially P2 as elastic response of the brain
* flow patterns (e.g. flow reversal in the aquaeduct, ...)
* amount of CSF outflow into the spinal cord




## Implementation challenges
* choice of framework for function spaces on different domains:
	* fenics mixed-dimensional branch
	* multiphenics
	* avoid the problem by formulations, that use the same functions on the whole domain
	* ignore the redundant degrees of freedom (drop zero entries)
* monolitic vs coupled iterative approach
* mesh movement necessary?
* preconditioning and robustness with iterative solvers
* proper scheme for NS solutions
